```python
#導入YOLO
from ultralytics import YOLO
model = YOLO('model/yolo11n.pt') #導入模型

#訓練
model.train(data = "config/dataset_custom.yaml",
            imgsz = 640,
            batch=8, #同時學習四張照片
            epochs = 100, #訓練100次
            workers = 1,
            device=0, #使用GPU 0
            project='Mango_training', 
            name='train1')
```
